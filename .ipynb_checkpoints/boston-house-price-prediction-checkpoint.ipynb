```python
#Import alll the libraries
import numpy as np
import pandas as pd
from sklearn import metrics
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Load the boston dataset
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
 #Initializing the dataframe
data = pd.DataFrame(boston.data)
```

```python
#Get the head of the Data frame
data.head()
```

```python
#adding the column name with labels
data.columns=boston.feature_names
data.head()
```

```python
#Adding target variable into the dataframe
data['PRICE'] = boston.target 
```

```python
#Now check the shape of the dateframe gives information about Metrics details like(row * column)
data.shape
```

```python
# Plotting the heatmap of correlation between features
plt.figure(figsize=(20,20))
sns.heatmap(corr, cbar=True, square= True, fmt='.1f', annot=True, annot_kws={'size':15}, cmap='Greens')
plt.savefig('pic1.jpg')
```

```python
# Spliting target(Y) variable and independent variables(X)
X = data.drop(['PRICE'], axis = 1)
Y = data['PRICE']
```

```python
#print independent feature data
X
```

```python
#print target variable data
Y
```

```python
#So let's jump into the model training and prediction part

#First step the split the data into training set and test set
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=4)

```

```python
X_train
```

```python
#Import the library for 

#Traning the model
from sklearn.linear_model import LinearRegression

lm=LinearRegression()
```

```python
#Then fit the model using training data
lm.fit(X_train,y_train)
```

```python
#Model prediction with test data

y_predict=lm.predict(X_test)
```

```python
y_predict
```

```python
#The final steps is to evaluate the model

#MAE : It is the mean of the absolute value of the errors. It measures the difference between two continuous variables, here actual and predicted values of y. 

#MSE: The mean square error (MSE) is just like the MAE, but squares the difference before summing them all instead of using the absolute value. 

#RMSE: The mean square error (MSE) is just like the MAE, but squares the difference before summing them all instead of using the absolute value. 

print('MAE:',metrics.mean_absolute_error(y_test, y_predict))
print('MSE:',metrics.mean_squared_error(y_test, y_predict))
print('RMSE:',np.sqrt(metrics.mean_squared_error(y_test, y_predict)))
```

```python
#Final Steps is to visualize the actual price and predicted price using matplotlib
color = np.arange(200)
plt.scatter(y_test, y_predict)
plt.xlabel("Prices")
plt.ylabel("Predicted prices")
plt.title("Prices vs Predicted prices")
plt.savefig('pedicted.jpg')
plt.show()

```
